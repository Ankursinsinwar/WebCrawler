```python
from email.message import EmailMessage
import smtplib
import mimetypes
import os
import requests
import pandas as pd
import ssl
import json
```

```python
question_list = [[],[]]

keyword= "MySQL interview question"
#keyword= input("Enter a keyword: ")
keyword1 = keyword

site_list=["quora.com","stackoverflow.com"]
for site in site_list:
  keyword = keyword +" site: "+site

  google_url = "Your google url with custom search api in it"
  google_url = google_url + "&q=" + keyword
  # print("Google URL: "+ google_url)

  response = requests.get(google_url)

  json_response = json.loads(response.text)

  try:
    for item in json_response ["items"]:
      title = item["title"]
      link = item["link"]
      title = title.replace(" Quora"or"Stack Overflow","")
      question_list[0].append(title)
      question_list[1].append(link)
  except Exception as e:
    print(e)

  keyword = keyword1
question_list
questions_dict = {"Questions": question_list[0],"links": question_list[1]}
df = pd.DataFrame(data=questions_dict)
df
```

```python
df.to_csv(keyword1+".csv")
```
